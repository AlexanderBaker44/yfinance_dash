```python
import random
nas=tickers_nasdaq()
random.shuffle(nas)
m=nas[0:1000]
```

```python

dfs = []
for i,j in enumerate(m):
    t = yf.Ticker(j)
    b = t.balance_sheet.T
    b['ticker']=[j]*len(b)
    dfs.append(b)
    if i%10==0:
        print(i)
        print(b)
```

```python
full_df = pd.concat(dfs,axis=0)
```

```python
full_df.to_csv('stocks.csv')
```

```python
na_df = full_df.isna().sum()
kept_cols = na_df[na_df<=25].index
```
